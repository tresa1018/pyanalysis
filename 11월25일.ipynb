```python
#막대그래프 함수 MakeBarChart()
def MakeBarChart(x,y,color, xlabel, ylabel, title):
    plt.figure(figsize=(10,5))
    plt.bar(x,y,color=color, alpha=0.7)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    
    YTICKS_INTERVAL=50000
    maxlim=(int(y.max()/YTICKS_INTERVAL)+1)*YTICKS_INTERVAL
    print(maxlim)
    values=np.arange(0, maxlim+1, YTICKS_INTERVAL)
    plt.yticks(values,['%s' %format(val,',')for val in values])
    
    #그래프 위에 건수와 비율 구하기
    ratio=100*y/y.sum()

    plt.rc('font',size=10) 
    for idx in range(y.size):
        value=format(y[idx],',')+'건'
        ratioval='%.1f%%' %(ratio[idx])
        #그래프 위에" 건수 표시
        plt.text(x=idx, y=y[idx]+1, s=value,horizontalalignment='center')
        #그래프의 중간에 비율 표시
        plt.text(x=idx, y=y[idx]/2, s=ratioval,horizontalalignment='center')
    
    #평균갑을 수평선으로 그리기
    meanval=y.mean()

    average='평균: %d건' %meanval
    plt.axhline(y=meanval, color='r', linewidth=1, linestyle='dashed')
    plt.text(x=y.size-1, y=meanval+200, s=average,horizontalalignment='center')
  
color=['b','g','r','c','m','y','k']
mycolor=color[0:len(chartdata)]

MakeBarChart(x=chartdata.index,y=chartdata, color=mycolor, 
             xlabel='국가명',ylabel='발생건수',title='국가별 코로나 발생건수(4월6일)')
```

```python
#누적형 막대그래프
def MakeBarChart01(chartdata, rotation, title, ylim=None, yticks_interval=50000):
    plt.figure(figsize=(10,5))
    chartdata.plot(kind='bar', rot=rotation, title=title, label=WHEN, stacked=True)
    plt.legend(loc='best')
    #국가별 누적 합을 최대값에 대한 연산으로 표현
    maxlim=(int(max(chartdata.sum(axis=1))/yticks_interval)+1)*yticks_interval
    values=np.arange(0, maxlim+1, yticks_interval)
    plt.yticks(values, ['%s' %format(val,',') for val in values])
 
    #y축의 상하한 값이 주어지는 경우에만 설정
    if ylim !=None:
        plt.ylim(ylim)
        
data.index.name='국가명'
data.columns.name='일자'

MakeBarChart01(chartdata=data, rotation=0, title='국가별 일별 발생 건수')
    
```

```python
filename = 'c:/users/LG/데이터시각화자료/data/tips.csv'
myframe=pd.read_csv(filename,index_col=0)
DINNER, LUNCH='Dinner','Lunch'

frame01=myframe.loc[myframe['time']==DINNER, 'total_bill']
frame01.index.name=DINNER

frame02=myframe.loc[myframe['time']==LUNCH, 'total_bill']
frame02.index.name=LUNCH

chartdata=[np.array(frame01),np.array(frame02)]

xtick_label=[DINNER, LUNCH]#x축을 위한 레이블
fig, (ax1,ax2)=plt.subplots(nrows=1, ncols=2, figsize=(9,4))

#상자수염 그래프 그리기
bplot1=ax1.boxplot(chartdata, vert=True, #세로 박스 설정
                    patch_artist=True, # 박스에 색 채우기
                    labels=xtick_label)
ax1.set_title('Rectangular box plot')

#notch 형태의 그래프
bplot2=ax2.boxplot(chartdata, notch=True, vert=True, 
                  patch_artist=True, labels=xtick_label)
ax2.set_title('Notched box plot')

#채우기 색넣기
colors=['pink','lightblue']
for bplot in (bplot1, bplot2):
    for patch, color in zip(bplot['boxes'],colors):
        patch.set_facecolor(color)
        
#수평선 그리기
for ax in [ax1, ax2]:
    ax.yaxis.grid(True)
    ax.set_xlabel('점심과 저녁 팁')
    ax.set_ylabel('observed data')   
```

```python
filename = 'c:/users/LG/데이터시각화자료/data/주요발생국가주간동향(4월2째주).csv'
data=pd.read_csv(filename, index_col='국가')
fig,ax=plt.subplots(figsize=(10,5), subplot_kw=dict(aspect='equal'))
COUNTRY=['독일','프랑스','중국','영국','이탈리아']
data=data.loc[COUNTRY,['4월06일']]

#도넛형 그리기
wedges,texts=ax.pie(data.values.flatten(), wedgeprops=dict(width=0.5), startangle=-40)

#연결선과 직사각형
bbox_props=dict(boxstyle="square, pad=0.3", fc="w", ec="k",lw=0.52)
kw=dict(arrowprops=dict(arrowstyle=" - "),
                       bbox=bbox_props,zorder=0, va="center")

for i, p in enumerate(wedges):
    ang=(p.theta2 - p.theta1)/2. +p.theta1
    y=np.sin(np.deg2rad(ang))
    x=np.cos(np.deg2rad(ang))
    horizontalalignment={-1:"right",1:"left"}[int(np.sign(x))]
    connectionstyle="angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle":connectionstyle})
    
    ax.annotate(COUNTRY[i],xy=(x,y), xytext=(1.3*np.sign(x), 1.4*y),
               horizontalalignment=horizontalalignment, **kw)
ax.set_title('국가별 4월 6일 주요 발생건수-도넛그래프')
```
