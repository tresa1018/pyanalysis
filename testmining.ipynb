```python
obama=open('Obama.txt',encoding='UTF-8').read() #Obama연설문 불러오기
obama
```

```python
import re            #불필요한 문자 제거를 위한 문자처리 패키지 불러오기 
obama=re.sub('[^가-힣]',' ',obama) #한글이 아닌 모든 문자를 의하는 규칙적용하여 제거하기(정규표현식)
obama
```

```python
#명사를 추출하기
import konlpy
hannanum=konlpy.tag.Hannanum()

nouns=hannanum.nouns(obama)
nouns
```

```python
import pandas as pd
df_word=pd.DataFrame({'word': nouns}) #데이터프레임으로 변환하기
df_word
```

```python
df_word['count']=df_word['word'].str.len()  #글자수 카운트하여 열 추가하기
df_word
```

```python
df_word=df_word.query('count > =2') #두글자 이상만 표현하기
df_word.sort_values('count')
```

```python
# 단어별 분리한 것을 빈도수에 따라 내림차순 정렬하기
df_word=df_word.groupby('word',as_index=False)\
.agg(n=('word','count'))\
.sort_values('n',ascending=False)     
df_word
```

```python
#단어빈도 막대 그래프만들기위한 상위20개만 추출
top20=df_word.head(20)
top20
```

```python
# 상위20개의 막대그래프
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams.update({'font.family' : 'Malgun Gothic',
                    'figure.dpi': '120',
                    'figure.figsize' : [6.5,6]})
ax=sns.barplot(data=top20, y='word',x='n')
ax.set_ylabel('단어')
ax.set_xlabel('단어수')
ax.set_title('Obama 대통령 당선 연설문')
```

```python
dic_word=df_word.set_index('word').to_dict()['n'] #위의 데이터프레임을 딕셔너리로 전환
dic_word
```

```python
#워드클라우드 모양만들기
from wordcloud import WordCloud

wc=WordCloud(random_state=1234,
            font_path = "c:/Windows/fonts/malgun.ttf",
            width=400,
            height=400,
            background_color='white')

#워드 클라우드 만들어 출력하기(테두리선 없애기)
img_wordcloud=wc.generate_from_frequencies(dic_word)
plt.figure(figsize=(5,5))
plt.axis('off')
plt.imshow(img_wordcloud)
```
